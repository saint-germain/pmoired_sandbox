```python
import pandas as pd
```

```python
data=pd.read_csv('HD163296_200_500.fullprobs',names=['a','b','c'])
#data=data[2:]
```

```python
plt.plot(data.a[:1000])
```

```python
n=200
nburn=100
cha=500
dcut=data[nburn:cha]
for i in range(1,n):
    dcut=pd.concat([dcut,data[nburn+i*cha:(i+1)*cha]],axis=0)
#    print(nburn+i*cha,(i+1)*cha)
```

```python
plt.plot((dcut.a[:1000]).to_numpy())
```

```python
dcut=dcut.to_numpy()
```
